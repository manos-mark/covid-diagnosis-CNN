### Imports

```python
import pandas as pd
import os
import shutil
import random
```

### Classification of samples (negative - positive)

```python
METADATA_PATH = r"C:\Users\Manos\Desktop\covid-chestxray-dataset\metadata.csv"
SAMPLES_PATH = r"C:\Users\Manos\Desktop\covid-chestxray-dataset\images"
PROJECTS_PATH = os.getcwd()
COVID_SAMPLES_PATH = PROJECTS_PATH+"\\Covid" 
NONCOVID_SAMPLES_PATH = PROJECTS_PATH+"\\NonCovid"

# Create a directories in which we will separate the samples
try:
    os.mkdir(COVID_SAMPLES_PATH)
    os.mkdir(NONCOVID_SAMPLES_PATH)
except OSError:
    print("The directories already exist")

# Load the metadata.csv which includes additional information about the 
metadata = pd.read_csv(METADATA_PATH)
```

#### Find the patients with COVID-19 , and their medical image is  posteroanterior (PA) view (of Frontview)

```python
covid_patients_selection = metadata.loc[(metadata.finding == "Pneumonia/Viral/COVID-19") & (metadata.view=="PA")]

for i in range(len(covid_patients_selection)):
    image_name = covid_patients_selection.iloc[i].filename
    try:
        shutil.copyfile(
            os.path.join(SAMPLES_PATH,image_name),
            os.path.join(COVID_SAMPLES_PATH,image_name)
        )
        print(f"The image {image_name} moved")
    except shutil.SameFileError:
        print(f"The file {image_name} already Exists")
    
    
```

### Get the normal patients (non infected with covid) using Kaggle dataset
Ref : https://www.kaggle.com/paultimothymooney/chest-xray-pneumonia  (train/Normal) 

```python
NORMAL_SAMPLES_PATH = r"C:\Users\Manos\Desktop\covid-chestxray-dataset\chest_xray\train\NORMAL" 
normal_image_names = os.listdir(NORMAL_SAMPLES_PATH)
random.shuffle(normal_image_names)
for i in range(len(os.listdir(COVID_SAMPLES_PATH))):
    try:
        shutil.copyfile(
            os.path.join(NORMAL_SAMPLES_PATH,normal_image_names[i]),
            os.path.join(NONCOVID_SAMPLES_PATH,normal_image_names[i])
        )
        print(f"The image {normal_image_names[i]} moved")
    except shutil.SameFileError:
        print(f"The file {normal_image_names[i]} already Exists")
        
```
